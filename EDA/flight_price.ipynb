```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.drop('Route', axis=1, inplace=True)
df.head()
```

```python
df['Airline'].value_counts()
```

```python
df['Additional_Info'].value_counts()
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df['Total_Stops'].value_counts()
```

```python
mapping_dict = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
}
df['Total_Stops'] = df['Total_Stops'].map(mapping_dict)
df['Total_Stops'].value_counts()
```

```python
df.info()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df = df.dropna(subset=['Total_Stops'])
```

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
df.head()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minute'] = df['Dep_Time'].str.split(':').str[1]
df.drop('Dep_Time', axis=1, inplace=True)
df.head()
```

```python
df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
df['Duration_Minute'] = df['Duration_Minute'].fillna(0)
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert specific columns to int
columns_to_convert = ['Arrival_Hour', 'Arrival_Minute', 'Dep_Hour', 'Dep_Minute', 'Duration_Hour', 'Duration_Minute']
df[columns_to_convert] = df[columns_to_convert].astype(int)
```

```python
df['Duration_Minute'] = df['Duration_Minute'].astype(int)
```
